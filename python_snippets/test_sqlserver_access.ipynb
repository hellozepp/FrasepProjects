```python
import swat
import os
swat.options.cas.print_messages = True
```

```python
os.environ["CAS_CLIENT_SSL_CA_LIST"] = "/opt/sas/viya/config/etc/SASSecurityCertificateFramework/cacerts/trustedcerts.pem"
cashost='frasepviya35smp'
casport=5570
conn = swat.CAS(cashost, casport)

conn.sessionProp.setSessOpt(metrics='TRUE', timeout=1800, caslib='casuser', messageLevel="ALL")
```

```python
conn.table.tableDetails(
    level="block",
    caslib="sqlsrv",
    name="prdsal2"
)
```

```python
# Read in the hmeq CSV to an in-memory data table and create a CAS table object reference
castbl = conn.read_csv(os.environ.get("HOME")+"/Courses/EVMLOPR/DATA/hmeq.csv", casout = dict(name="hmeq",caslib="sqlsrv", replace=True))
```

```python
castbl.save(name="hmeq",caslib="sqlsrv",replace="true")
```
